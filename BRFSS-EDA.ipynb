```python

# Extract the 6 columns of interest
cols = ['_STATE', 'INCOME3', 'PRIMINS2', '_DENVST3', 'LASTDEN4', 'RMVTETH4', '_LLCPWT']
df = brfss[cols].copy()

# --- Value mappings from BRFSS 2024 codebook ---
# (Don't know / Not sure / Refused codes are excluded — they become NaN and are dropped later)

state_map = {
    1: 'Alabama', 2: 'Alaska', 4: 'Arizona', 5: 'Arkansas', 6: 'California',
    8: 'Colorado', 9: 'Connecticut', 10: 'Delaware', 11: 'District of Columbia',
    12: 'Florida', 13: 'Georgia', 15: 'Hawaii', 16: 'Idaho', 17: 'Illinois',
    18: 'Indiana', 19: 'Iowa', 20: 'Kansas', 21: 'Kentucky', 22: 'Louisiana',
    23: 'Maine', 24: 'Maryland', 25: 'Massachusetts', 26: 'Michigan',
    27: 'Minnesota', 28: 'Mississippi', 29: 'Missouri', 30: 'Montana',
    31: 'Nebraska', 32: 'Nevada', 33: 'New Hampshire', 34: 'New Jersey',
    35: 'New Mexico', 36: 'New York', 37: 'North Carolina', 38: 'North Dakota',
    39: 'Ohio', 40: 'Oklahoma', 41: 'Oregon', 42: 'Pennsylvania',
    44: 'Rhode Island', 45: 'South Carolina', 46: 'South Dakota',
    48: 'Texas', 49: 'Utah', 50: 'Vermont', 51: 'Virginia',
    53: 'Washington', 54: 'West Virginia', 55: 'Wisconsin', 56: 'Wyoming',
    66: 'Guam', 72: 'Puerto Rico', 78: 'Virgin Islands'
}

income_map = {
    1:  'Less than $10,000',
    2:  '$10,000–<$15,000',
    3:  '$15,000–<$20,000',
    4:  '$20,000–<$25,000',
    5:  '$25,000–<$35,000',
    6:  '$35,000–<$50,000',
    7:  '$50,000–<$75,000',
    8:  '$75,000–<$100,000',
    9:  '$100,000–<$150,000',
    10: '$150,000–<$200,000',
    11: '$200,000 or more',
}

primins_map = {
    1:  'Employer',
    2:  'Private/Non-employer',
    3:  'Medicare',
    4:  'Medigap',
    5:  'Medicaid',
    6:  'CHIP',
    7:  'Military/VA',
    8:  'Indian Health Service',
    9:  'State plan',
    10: 'Other government',
    88: 'No coverage',
}

denvst_map = {
    1: 'Yes',
    2: 'No',
}

lastden_map = {
    1: '<12 months',
    2: '1–<2 years',
    3: '2–<5 years',
    4: '5+ years',
    8: 'Never',
}

rmvteth_map = {
    1: '1 to 5',
    2: '6 or more',
    3: 'All',
    8: 'None',
}

# Apply mappings
df['_STATE']   = df['_STATE'].map(state_map)
df['INCOME3']  = df['INCOME3'].map(income_map)
df['PRIMINS2'] = df['PRIMINS2'].map(primins_map)
df['_DENVST3'] = df['_DENVST3'].map(denvst_map)
df['LASTDEN4'] = df['LASTDEN4'].map(lastden_map)
df['RMVTETH4'] = df['RMVTETH4'].map(rmvteth_map)

df.head(10)
```

```python
df = df.dropna()
print(f"Rows remaining: {len(df):,}")
df.head(10)
```

```python
# Metric 1: Weighted % who visited dentist in past year (_DENVST3 = Yes)

m1 = (brfss[['_STATE', '_DENVST3', '_LLCPWT']]
      .copy()
      .assign(_STATE   = lambda x: x['_STATE'].map(state_map),
              _DENVST3 = lambda x: x['_DENVST3'].map(denvst_map))
      .dropna(subset=['_STATE', '_DENVST3', '_LLCPWT'])          # keep only Yes/No with valid weight
)

metric1 = (
    m1.groupby(['_STATE', '_DENVST3'])['_LLCPWT'].sum()
      .unstack('_DENVST3')                                        # columns: Yes | No
      .assign(pct_visited = lambda x: x['Yes'] / (x['Yes'] + x['No']) * 100)
      [['Yes', 'No', 'pct_visited']]
      .rename(columns={'Yes': 'weight_yes', 'No': 'weight_no',
                       'pct_visited': '% Visited Dentist (Past Year)'})
      .round(2)
      .sort_values('% Visited Dentist (Past Year)', ascending=False)
)

display(metric1)
```
